```python
import os, sys
sys.path.append("../")
from main import sigmoid, tanh, relu, leaky_relu, softmax
import numpy as np
import matplotlib.pyplot as plt
```

```python
Sig = []
X = np.linspace(-5, 5, 100)

CST = len(X)*[0.5]
print(CST)

for i in X:
    Sig.append(sigmoid(i))

plt.xticks(np.arange(-5, 6, step=1))
plt.plot(X, Sig, label='Sigmoid Function')
plt.plot(X, CST, 'r-')
plt.title("Sigmoid function")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.axvline(x=0, color='r')

plt.grid()
plt.savefig("./sigmoid_function.png")
plt.show()
```

```python
Tanh = []
X = np.linspace(-5, 5, 100)

CST = len(X)*[0]

for i in X:
    Tanh.append(tanh(i))

plt.xticks(np.arange(-5, 6, step=1))
plt.plot(X, Tanh, label='Tanh Function')
plt.plot(X, CST, 'r-')
plt.title("Tanh function")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.axvline(x=0, color='r')

plt.grid()
plt.savefig("./tanh_function.png")
plt.show()
```

```python
Relu = []
X = np.linspace(-5, 5, 100)

CST = len(X)*[0]

for i in X:
    Relu.append(relu(i))

plt.xticks(np.arange(-5, 6, step=1))
plt.plot(X, Relu, label='ReLU Function')
plt.title("ReLU function")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.axvline(x=0, color='r')

plt.grid()
plt.savefig("./relu_function.png")
plt.show()
```
